```python
from pathlib import Path
import pandas as pd

# Define the directory and pattern
directory = Path('output_training/phase1_dataset1')
pattern = '*_performance.csv'

# List all matching files
csv_files = list(directory.glob(pattern))
classifer_names = [file.stem.split('_')[0] for file in csv_files]
# Read all CSV files and concatenate them into a single DataFrame

dataframes = [pd.read_csv(file) for file in csv_files]
df = pd.concat(dataframes, ignore_index=True)
df['classifer'] = classifer_names
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_theme(style="whitegrid")

# Create figure
plt.figure(figsize=(10, 6))

# Transposed barplot (swap x and y)
sns.barplot(
    y="imbalance_handling", 
    x="auc_pr", 
    hue="classifer", 
    data=df, 
    palette="Set2",  # Colorblind-friendly
    edgecolor="black", 
    linewidth=1.2
)

# Titles and labels
#plt.title("AUC-PR by Classifier and Imbalance Handling Method", fontsize=14, fontweight="bold")
plt.ylabel("Imbalance Handling Method", fontsize=12)
plt.xlabel("AUC-PR", fontsize=12)

# Adjust legend (move outside the plot)
plt.legend(title="Classifier", fontsize=10, title_fontsize="11", loc="upper left", bbox_to_anchor=(1, 1), frameon=True)

# Improve y-ticks
plt.yticks(fontsize=10)

# Layout adjustment
plt.tight_layout()

plt.savefig("phase1_auc_pr_plot.pdf", format="pdf", bbox_inches="tight")

# Show plot
plt.show()
```
