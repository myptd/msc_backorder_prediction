```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the directory and pattern
directory = Path('output_training/eval_dnn_phase2_dataset1_smote')
pattern = '*_performance.csv'

# Define the learning rates
learning_rates = ['0.001', '0.0001', '0.0005', '1e-05', '1e-06', '5e-05', '5e-06']

# Initialize a list to store data
data = []

# List all matching files
csv_files = list(directory.glob(pattern))

# Loop through each file and extract relevant data
for file in csv_files:
    # Extract classifier and learning rate from the filename
    file_name_parts = file.stem.split('_')
    classifier = file_name_parts[0] 
    learning_rate = file_name_parts[1]

    # Check if the learning rate is in the defined list
    if learning_rate in learning_rates:
        # Read the CSV file
        df = pd.read_csv(file)

        # Extract AUC-ROC value (assuming it's a column in the CSV)
        auc_roc = df['auc_roc'].values[0]  # Assuming AUC-ROC is in the first row

        # Append data to the list
        data.append({'learning_rate': learning_rate, 'classifier': classifier, 'auc_roc': auc_roc})

# Create a DataFrame from the collected data
df_plot = pd.DataFrame(data)

# Convert 'learning_rate' to categorical for proper ordering
df_plot['learning_rate'] = pd.Categorical(df_plot['learning_rate'], categories=learning_rates, ordered=True)

# Plotting the AUC-ROC for each learning rate and classifier
plt.figure(figsize=(14, 8))
sns.barplot(x='learning_rate', y='auc_roc', hue='classifier', data=df_plot)
#plt.title('AUC-ROC by Learning Rate and Classifier')
plt.xlabel('Learning Rate')
plt.ylabel('AUC-ROC')
plt.legend(title='Classifier')
plt.ylim(0, 1)  # Set y-axis limits to 0-1 for AUC-ROC
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("phase3_auc_roc_plot.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
